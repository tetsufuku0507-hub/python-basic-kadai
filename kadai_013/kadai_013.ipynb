```python
def calculate_tax_included(price: int, tax_rate: int) -> int:
    """商品の金額と消費税率から、税込価格を計算する関数"""
    # 消費税額を計算 (金額 * 税率 / 100)
    # 税込金額 = 金額 + 消費税額
    tax_included_price = price * (1 + tax_rate / 100)

    # 戻り値を整数（int）に変換して返す
    return int(tax_included_price)

# --- 関数の呼び出し ---
# 第1引数に金額「110」、第2引数に消費税「10」を設定
result = calculate_tax_included(110, 10)

print(f"商品の税込金額は {result} 円です。")
```
