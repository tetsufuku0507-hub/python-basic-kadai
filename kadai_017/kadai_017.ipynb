```python
class Human:
    # クラス：初期化メソッドで属性（名前・年齢）を設定
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 関数（メソッド）：大人かどうかを判定して標準出力する
    def check_adult(self):
        # 条件分岐：20歳以上かどうか
        if self.age >= 20:
            print(f"{self.name}さんは、{self.age}歳なので大人である。")
        else:
            print(f"{self.name}さんは、{self.age}歳なので大人でない。")

# ディクショナリ：初期データを管理（可読性を高めるため）
user_data = [
    {"name": "侍 太郎", "age": 25},
    {"name": "侍 花子", "age": 18},
    {"name": "侍 二郎", "age": 30},
    {"name": "侍 桃子", "age": 12}
]

# リスト：Humanクラスのインスタンスを格納する
human_list = []

# 繰り返し処理：データを元にインスタンスを生成し、リストに追加
for data in user_data:
    # 変数のスコープ：変数humanはこのループ内（およびその後の参照）で有効
    human = Human(data["name"], data["age"])
    human_list.append(human)

# 繰り返し処理：リストの要素数分だけメソッドを呼び出す
print("--- 判定結果 ---")
for person in human_list:
    person.check_adult()
```
