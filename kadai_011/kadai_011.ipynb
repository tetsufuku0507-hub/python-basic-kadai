```python
# 対象リスト
array = ["水","金","地","火","木","土","天","海","冥"]

# --- forループによる出力 ---
# 先頭からリスト内の要素を直接ひとつずつ取り出す
print("--- forループによる出力 ---")
for planet in array:
    print(planet)

# 空行を挟む（見やすくするため）
print()

# --- whileループによる出力 ---
# インデックス（添字）を0から始め、リストの長さ未満の間繰り返します
print("--- whileループの結果 ---")
index = 0
while index < len(array):
    print(array[index])
    index += 1  # インデックスを1ずつ増やして次の要素へ
```
