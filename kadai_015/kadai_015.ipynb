```python
class Human:
    """人間を表現するクラス"""

    def __init__(self, name, age):
        """
        コンストラクタ（インスタンス化の際に実行される）
        :param name: 名前（属性）
        :param age: 年齢（属性）
        """
        # 属性(name, age)に値を代入
        self.name = name
        self.age = age

    def printinfo(self):
        """
        属性を表示するメソッド
        """
        print(f"名前: {self.name}, 年齢: {self.age}歳")

# --- クラスの利用（メイン処理） ---

# 1. Humanクラスのインスタンスを作成し、変数に代入（インスタンス化）
person1 = Human("田中 太郎", 25)
person2 = Human("佐藤 花子", 30)

# 2. メソッドを呼び出して情報を出力
person1.printinfo()
person2.printinfo()

# 3. 直接属性にアクセスして利用することも可能
print(f"{person1.name}さんの来年の年齢は {person1.age + 1}歳です。")
```
